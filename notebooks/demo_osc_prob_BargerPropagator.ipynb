```python
from wrapper_Prob3 import BargerPropagator


PREM_file = "../data/PREM.dat"

BP = BargerPropagator.PyBargerPropagator(PREM_file)
```

```python
t12   = 0.307
t13   = 0.022
t23   = 0.58
dm21  = 7.53e-5
mAtm  = 2.4e-3
delta = 4.53

ks = 1   
prod_height = 20.

# BP.definePath(-0.9, prod_height, True)
# BP.setMNS(t12, t13, t23, dm21, mAtm, delta, 1000, ks, 1)
```

```python
def get_osc_prob(BP, energy, cosZ, nutype, flav_ini, flav_end):
    BP.definePath(cosZ, prod_height, True)
    BP.setMNS(t12, t13, t23, dm21, mAtm, delta, energy, ks, nutype)
    BP.propagate(nutype)
    return BP.getProb(flav_ini, flav_end)
```

```python
import numpy as np

cos_Z = -0.9
nutype = 1
flavor_A = 1 # mu
flavor_B = 1 # mu

p = []
energies = np.linspace(0,1, 10000)
for E in energies:
    p.append(get_osc_prob(BP, E, cos_Z, nutype, flavor_A, flavor_B))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(energies, p)
plt.gca().set_ylabel('Osc Prob')
plt.gca().set_xlabel('Energy [Gev]')
```
